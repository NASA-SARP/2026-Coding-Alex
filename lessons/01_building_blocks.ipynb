## `if` statements

Often after a comparison we want to make a decision based on if the comparison was `True` or `False`.  For example, 
> if the cloud cover was less than 30 percent, start the processing.  Otherwise, skip to the next image.

We do that in Python with `if` statements, which have the following syntax:

```
if <<True/False statement>>:
    <<do something>>
else:
    <<do something else>>
```
Statements contained in `<<` `>>` are placeholders and should be filed with python code.  The `if`, `:`, `else:` and indenting are all part of the required syntax for an if statement.

If you have multiple statements you can add in as many as you need after the first `if` using the syntax `elif`. However, note that an `if`  statement will exit as soon as it finds the first statement that returns `True`, starting from the top.

````python
# If statement example
cloud_cover = 50
if cloud_cover < 30:
    print('Good image for processing')
else:
    print('Too many clouds for processing')
````

````python
# Including elif
cloud_cover = 35
if cloud_cover < 30:
    print('Good image for processing')
elif cloud_cover < 40:
    print('Potentially usable')
elif cloud_cover < 50:
    print('Only if we are really desperate')
else:
    print('Too many clouds for processing')
````

## Bonus concept: f-strings

We won't practice these for now, but since we've just covered `print` statements and variables, it's worth mentioning one of my favorite handy tools for formatting things that you might want to print: formatted string literals, or f-strings.

The main idea to is to preface our string with the letter `f` and use what we call replacement fields `{}` inside the string literal where we want to insert variables. After a bit of practice, you will quickly appreciate the flexibility that f-strings provided for printing statements that include multiple data types in a tidy format.

You can read more about f-strings here: https://realpython.com/python-string-formatting/

# Data Structures: Lists and Dictionaries

So far we have looked at variables, data types, operators, and `if` statements. These represent the first level of coding building blocks.

## Lists
**Lists** are a data structure which represent a group of values. Their syntax is square brackets `[]` with a comma in between each item. The items can be any data types, and we can also mix data types in a single list. Assigning a list works the same as assigning a variable using `=`.

````python
# List of pollutants - strings
['co', 'co2', 'no2', 'o3']
````

````python
# List of pH values - floats
[4.5, 1.0, 4.3, 4.3, 7.6]
````

````python
list1 = ["apple", "banana", "cherry"]
list2 = [1, 5, 7, 9, 3]
list3 = [True, False, False]
list4 = ["abc", 34, True, 40, "male"]
print(list4)
````

## Indexing lists

You access individual items in a list using an **index**.  An index is a number which represents the position of an item in the list. __Indexing__ as a verb refers to getting the value of that item using its index.

In Python (and most programming langauges) you begin counting with indexes at 0 instead of at 1.  So, to get the first item, we need to use index 0. This concept of starting counting at 0 instead of 1 is true across the whole Python language.

````python
colors = ['red', 'green', 'blue', 'yellow', 'white', 'black']
````

````python
# Which pollutant is at position 1 in the `pollutants` list?
print('First item:', colors[0])
print('Second item:', colors[1])
print('Last item:', colors[-1])
print('Second last item:', colors[-2])
````

To access several values consecutively you use a colon `:`.

The general syntax is:
> `list[start:end+1]`

The fact that you need `end+1` can feel a little off. The benefit of this approach is that subtractding your `start` from your `end` gives the number of items that should be returned. So in the example above, 3-1 = 2 and 2 items were returned from the list.

````python
print(colors[1:3])
print(len(colors[1:3])) # we will come back to len() shortly!
````

## Doing things to lists: Functions/Methods and Arguments

This is where things start to get fun!

If data structures are like nouns in the english language then methods or functions are the verbs. Thus far we have created lots of variables of many different types.  **Functions** or **methods** will now enable us to do things to them. On a vocabulary note, for now you can consider methods and functions to be the same thing.

We won't get too hung up on them for now, but it is good to mention here that **arguments** are the other key concept that tends to accompany functions. If functions are like verbs in the english language, then arguments are like adverbs: they add detail that tells the computer _how_ to perform a certain function operation. Note that arguments can be a plain argument or a keyword argument (kwarg), but more on this in later lessons.

Here are some of the most __common functions that are relevant to lists:__

| Function / Method   | Description | Note about arguments |
|:---|:---|:---|
| `len()` |  returns the length of the list (number of items in it)| argument is the object or data structure (e.g., list) we want to know the length of|
| `sort()` |  sorts the list in ascending order| `reverse=True` to sort in reverse order, `key=len` to sort by length|
| `append()` |  adds an item to the list| argument is the object we want to add|
| `extend()` |  adds multiple items to the list| argument is the objects we want to add|
| `insert()` |  add an item at a specific position in the list| arguments are the position and the object we want to add|
| `remove()` |  removes items from the list| argument is the object we want to remove|
| `pop()` |  removes an item from the list by its index position| argument is the index of the item we want to remove|
| `count()` |  counts the number of items that equal the argument| argument is an object you want to count in the list|


You can read more about the many functions and methods for workking with lists: https://towardsdatascience.com/python-lists-the-definitive-guide-for-working-with-ordered-collections-of-data-53b06a194826/

````python
numbers = [5, 3, 6, 2, 1]
len(numbers)

numbers.append(9)
print(numbers)

numbers.extend([10, 11, 12])
print(numbers)

numbers.sort()
print(numbers)

numbers.sort(reverse=True)
print(numbers)
````

````python
pollutants = ['co', 'co2', 'no2', 'o3']
len(pollutants)

# Appending methane
pollutants.append('ch4')
# Updated list
print(pollutants)

pollutants.sort()
print(pollutants)

pollutants.remove('co')
print(pollutants)

pollutants.pop(2)
print(pollutants)
````

## Updating a value in a list

Another thing you might want to take is changing the value of an item.  This is so common you don't need to use a `.verb()`-type function for this.  You just use the same `[]` notation we used for indexing and an equals sign to update the value.

````python
pollutants = ['co', 'co2', 'no2', 'o3']
print(pollutants)

# Changing the value
pollutants[2] = 'ch4'

# Notice the value in the 3rd position (index 2) has changed.
pollutants
````

## Dictionaries

Dictionaries are another data structure.  They are similar to lists in that they represent a group of things, but with dictionaries each value is given a name, called a **key**.

dictionary items have two parts - the **key** and the **value**. Each item in the dictionary, called a **key-value pair**, is seperated by a comma. The general syntax of a dictionary is therefore:
```
{key1: value1,
key2: value2,
key3: value3}
```

This concept is similar to a book dictionary, with words and definitions.  In that case the words are the keys and the definitions are the values.

The values in the dictionary can be any of the types of data types we know so far -- strings, numbers, booleans, lists, or even another dictionary.  The keys can be strings, numbers, or booleans, but not lists.

````python
# Example dictionary
human = {
        'name': 'Dyamond',
        'age': 25,
        'job': 'Dev',
        'city': 'New York',
        'email': 'dyamond@web.com',
        }

human
````

## Inspecting and modifying a dictionary

You can view a particular value of a dictionary using its key and square brackets `[]`.

If instead of a single value you want to view all either all of the keys or all of the values, you use `.keys()` and `.values()`, respectively.

You change or add a new dictionary item with the same `[]` syntax you use to access the item's value and a `=` (just like with a list). This syntax works whether or not the dictionary key exists already or not. If the key does not exist, then this will add a new key.

````python
metadata = {
    'location': 'Nairobi', 
    'quality_flag': 5, 
    'zenith': 60, 
    'clouds': True,
    'data_center': ['LPDAAC', 'ASDC'],
}

metadata['location']
````

````python
metadata.keys()
````

````python
metadata.values()
````

````python
print('value before', metadata['zenith'])
metadata['zenith'] = 30
print('value after', metadata['zenith'])
````

````python
print('value before:', metadata['agency'])
````

# File formats: JSON

We have learned about and seen a few examples of Python dictionaries. But how might you go about storing this kind of data structure if you wanted to use it again later? 

Dictionaries are often stored in a file format known as JSON (or JavaScript Object Notation). Dictionaries map naturally to JSON format because both use key–value pairs. JSON is a text-based data format and therefore has the benefit of being human readable (as well as computer readable).

So if we wanted to save our earlier `metadata` dictionary to a JSON, we could use:
```
metadata = {
    'location': 'Nairobi', 
    'quality_flag': 5, 
    'zenith': 60, 
    'clouds': True,
    'data_center': ['LPDAAC', 'ASDC'],
}

import json # Importing the package libraray

# Save dictionary to a JSON file using json.dump()
with open('metadata.json', 'w') as f:
    json.dump(metadata, f, indent=4)

# To read back into Python later using json.load()
with open('metadata.json', 'r') as f:
    metadata_loaded = json.load(f)

```
Try it out for yourself if you want to! The resulting JSON file should look something like (for example if you opened it in a text editor):

`{"location": "Nairobi", "quality_flag": 5, "zenith": 60, "clouds": true, "data_center": ["LPDAAC", "ASDC"]}`

Note that the JSON object requires the `"` quotations while a Python dictionary may use either `'` or `"`. `json.dump()` and `json.load()` automatically handle this conversion for us.
