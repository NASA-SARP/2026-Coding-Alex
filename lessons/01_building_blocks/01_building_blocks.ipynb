## Variables and data types

As programming tasks get more complex we want to be able to keep track of many values at once.  We do this is by assigning a name to values we want to keep track of. When we give a value a name it becomes a **variable**. Variables can have different **data types** such as: `integer`, `float`, `string`, and `bool`.

Variables often change values throughout a program as we operate on them. In this sense, it is important to understand the order of execution: the idea that **code is executed from top to bottom.**

| Python Data Type | Plain English  | Example |
|:---|:---:|:---:|
| `string` (`str`)| text | 'Tuesday' or "Tuesday" |
|`integer` (`int`)|  whole number  | 67 |

|  `boolean` (`bool`)| yes/no  | True or False |

## if statements

We often want to make a decision based on whether a comparison was `True` or `False`. We do this in Python with `if` statements. Multiple statements can be added after the first `if` using the syntax `elif` (although `elif` is not required). However, note that an `if`  statement will exit as soon as it finds the first statement that returns `True`, starting from the top.

```
if <<True/False statement>>:
    <<do something>>
elif <<True/False statement>>:
    <<do something>>
else:
    <<do something else>>
```



````python
# If statement example
cloud_cover = 50
if cloud_cover < 30:
    print('Good image for processing')
else:
    print('Too many clouds for processing')
````

````python
# Including elif
cloud_cover = 35 #60
if cloud_cover < 30:
    print('Good image for processing')
elif cloud_cover < 40:
    print('Potentially usable')
elif cloud_cover < 50:
    print('Only if we are really desperate')
else:
    print('Too many clouds for processing')
````

## Data Structures

The variables, data types, operators, and if statements represent the first level of coding building blocks. Combining these building blocks, we can create increasingly more complex objects which we call __data structures__. Data structures are made up of groups of data types, organized differently depending on the structure. Here, we will look at two types of data structure: **lists** and **dictionaries**.

## Lists
**Lists** are a data structure which represent a group of values. Their syntax is square brackets `[]` with a comma in between each item. The items can be any data types and we can also mix data types in a single list. Assigning a list works the same as assigning a variable using `=`.

````python
# List of pollutants - strings
['co', 'co2', 'no2', 'o3']
````

````python
# List of pH values - floats
[4.5, 1.0, 4.3, 4.3, 7.6]
````

````python
# list of bools
[True, False, False]
````

````python
# mixed list
mixed_list = ["abc", 34, True, 40, "male"]
print(mixed_list)
````

## Indexing and slicing lists

We access individual items in a list using an **index**: a number which represents the position of an item in the list. In Python (and most programming langauges) we begin counting with indexes at 0 instead of at 1, so to get the first item we use index 0.

````python
# Which pollutant is at position 1 in the `pollutants` list?
colors = ['red', 'green', 'blue', 'yellow', 'white', 'black']

print('First item:', colors[0])
print('Second item:', colors[1])
print('Last item:', colors[-1]) # negative indexing starts from -1 for the last item
print('Second last item:', colors[-2])
````

To access several values consecutively you use a colon `:`. This is known as **slicing**:
> `list[start:end+1]`

Note how we need `end+1`, which feels a little unintuitive. The benefit is that subtractding `start` from `end` gives the number of items.

There are a whole load of other ways you can index lists that we won't cover here (see further reading at the end of this lesson!)

````python
print(colors[1:3])
print(len(colors[1:3])) # we will come back to len() shortly!
````

:::{admonition} Vocabulary
__Indexing:__ accessing an item using its index position.

__Slicing:__ accessing multiple items using their range of index positions.

## Doing things to lists: Functions/methods and arguments

If data structures are like nouns then functions are the verbs. **Functions** (or **methods**, which we'll consider the same things for now) enable us to do things to the variables and data structures we've seen. 

We won't get hung up on them for now, but note that **arguments** are the other component that tends to accompany functions. If functions are like verbs, then arguments are like adverbs: they tell the computer specifics about _how_ to perform function operations. 

### List functions/methods

Here are some of the most common __functions for lists:__

| Function / Method   | Description | Note about arguments |
|:---|:---|:---|
| `len()` |  returns the length of the list (number of items in it)| argument is the object or data structure we want to know the length of|
| `sort()` |  sorts the list in ascending order| `reverse=True` to sort in reverse order, `key=len` to sort by length|
| `append()` |  adds an item to the list| argument is the object we want to add|
| `extend()` |  adds multiple items to the list| argument is the objects we want to add|
| `insert()` |  add an item at a specific position in the list| arguments are the position and the object we want to add|
| `remove()` |  removes items from the list| argument is the object we want to remove|
| `pop()` |  removes an item from the list by its index position| argument is the index of the item we want to remove|
| `count()` |  counts the number of items that equal the argument| argument is an object you want to count in the list|

````python
numbers = [5, 3, 6, 2, 1]
len(numbers)

numbers.append(9)
print(numbers)

numbers.extend([12, 11, 10])
print(numbers)

numbers.sort()
print(numbers)

numbers.sort(reverse=True)
print(numbers)
````

````python
pollutants = ['co', 'co2', 'no2', 'o3']
len(pollutants)

# Appending methane
pollutants.append('ch4')
# Updated list
print(pollutants)

pollutants.sort()
print(pollutants)

pollutants.remove('co')
print(pollutants)

pollutants.pop(2)
print(pollutants)
````

## Updating a value in a list

Changing the value of an item in a list is such a common task we don't need to use a `.verb()`-type function for this.  We just use the same `[]` notation we used for indexing and an equals sign to update the value.

````python
pollutants = ['co', 'co2', 'no2', 'o3']
print(pollutants)

# Changing the value
pollutants[2] = 'ch4'

# Notice the value in the 3rd position (index 2) has changed.
pollutants
````

## Dictionaries

Dictionaries are another data structure that are similar to lists (they represent a group of things), but with dictionaries each value is given a name, called a **key**. In turn, each key has a **value**. In this way, each item in the dictionary is called a **key-value pair**, which are seperated by a comma. This concept is similar to a book dictionary, with words and definitions.

Values can be any of the data types we've covered so far: strings, numbers, booleans, lists, or even another dictionary. Keys can be strings, numbers, or booleans, but not lists.

```
{key1: value1,
key2: value2,
key3: value3}
```

## Inspecting and modifying a dictionary

We can view a value of a dictionary using its key and square brackets `[]`. We can also view either all of the keys or all of the values using `.keys()` and `.values()`. We change or add a new dictionary item with the same `[]` syntax we use to access the item's value and a `=` (just like with a list). This syntax works whether or not the dictionary key exists already or not.

````python
metadata = {
    'location': 'Nairobi', 
    'quality_flag': 5, 
    'zenith': 60, 
    'clouds': True,
    'data_center': ['LPDAAC', 'ASDC'],
}

metadata
````

````python
# Access the value of the location key
metadata['location']
````

````python
# Get all the keys
metadata.keys()
````

````python
# Get all the values
metadata.values()
````

````python
# Update the value of the zenith key
print('value before', metadata['zenith'])
metadata['zenith'] = 30
print('value after', metadata['zenith'])
````

````python
print('value before:', metadata['agency'])
````

# File formats: JSON

Having seen a few examples of dictionaries, how might you go about storing this kind of data structure if you wanted to use it again later? 

Dictionaries are often stored in a file format known as JSON (or JavaScript Object Notation). Dictionaries map naturally to JSON format because both use key–value pairs. JSON is a text-based data format and therefore has the benefit of being human readable (as well as computer readable).

If we wanted to save our `metadata` dictionary to a JSON file, we could use:
```
metadata = {
    'location': 'Nairobi', 
    'quality_flag': 5, 
    'zenith': 60, 
    'clouds': True,
    'data_center': ['LPDAAC', 'ASDC'],
}

import json # Importing the package libraray

# Save dictionary to a JSON file using json.dump()
with open('metadata.json', 'w') as f:
    json.dump(metadata, f, indent=4)

# To read back into Python later using json.load()
with open('metadata.json', 'r') as f:
    metadata_loaded = json.load(f)

```
Try it out for yourself if you want to! The resulting JSON file should look something like this (for example if you opened it in a text editor):

`{"location": "Nairobi", "quality_flag": 5, "zenith": 60, "clouds": true, "data_center": ["LPDAAC", "ASDC"]}`

Note that the JSON object requires the `"` quotations while a Python dictionary may use either `'` or `"`. `json.dump()` and `json.load()` automatically handle this conversion for us.
