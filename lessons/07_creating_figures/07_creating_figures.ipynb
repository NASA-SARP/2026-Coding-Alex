## Figure and axes setup with matplotlib


```python
import matplotlib.pyplot as plt
```

```python
# Create figure and axes objects
fig = plt.figure()
ax = plt.axes()

print(type(fig))
print(type(ax))
print(ax)
```

```python
# Try commenting out either of the last two lines to see how the figure changes
fig = plt.figure()
ax = plt.axes()
fig.set_facecolor('orange')  # changing the color of the figure
ax.set_facecolor('cyan')  # changing the color of the axes
```

## Plotting real data

```python
# Import libraries
import pandas as pd
import numpy as np

# Import example data - flight from NAAMES campaign
naames = pd.read_csv('./data/naames-mrg01-c130_merge_20151112_R5_thru20151114.csv', skiprows=223)
# Filter down to just 1 day
naames = naames[naames[' Fractional_Day'] < 317]

# Remove NaN values
naames = naames.replace(-999999, np.nan)
```
